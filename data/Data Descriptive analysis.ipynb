## Descriptive Data Analysis

```python
import matplotlib.pyplot as plt 

#Easting Vs Northing
marker_size=15
plt.scatter(initdata['Easting'], initdata['Northing'], marker_size)
plt.title("Fig 1: Easting vs Northing")
plt.xlabel("Easting")
plt.ylabel("Northing")
cbar= plt.colorbar()
cbar.set_label("elevation (m)", labelpad=+1)
plt.show()
```

```python
#Easting Vs 1st Road Class
marker_size=15
plt.scatter(initdata['1st Road Class'], initdata['Easting'],marker_size)
plt.title("Fig 2: Easting Vs Roadclass")
plt.xlabel("1st Road Class")
plt.ylabel("Easting")
cbar= plt.colorbar()
cbar.set_label("elevation (m)", labelpad=+1)
plt.show()

#Northing Vs 1st Road Class
marker_size=15
plt.scatter(initdata['1st Road Class'], initdata['Northing'],marker_size)
plt.title("Fig 3: Northing Vs Road class")
plt.xlabel("1st Road Class")
plt.ylabel("Northing")
cbar= plt.colorbar()
cbar.set_label("elevation (m)", labelpad=+1)
plt.show()
```

```python
#Easting Vs 1st Road Class
marker_size=15
plt.scatter(initdata['Road Surface'], initdata['Easting'],marker_size)
plt.title("Fig 4: Road surface Vs Easting")
plt.xlabel("Road Surface")
plt.ylabel("Easting")
cbar= plt.colorbar()
cbar.set_label("elevation (m)", labelpad=+1)
plt.show()

#Northing Vs 1st Road Class
marker_size=15
plt.scatter(initdata['Road Surface'], initdata['Northing'],marker_size)
plt.title("Fig 5: Road surface Vs Northing")
plt.xlabel("Road Surface")
plt.ylabel("Northing")
cbar= plt.colorbar()
cbar.set_label("elevation (m)", labelpad=+1)
plt.show()
```

```python
import matplotlib.pyplot as plt

#Vehicles Vs Accident sevearities
y=initdata['Casualty Severity']=='Slight'
x=initdata['Number of Vehicles']

plt.scatter(x, y)
plt.title('Vehicles Vs Accident severity')
plt.xlabel('No. of vehicle')
plt.ylabel('Accident severity')
plt.show()
```

***********Inferences Derived from the attribute Easting and Northing **********

1)Figure 1 describes about the Northing and Easting data points in the dataset where the accident occured. From the above figures it explicitly knows that most of the accidents occured in the Easting: 430000-44000 and Northing: 425000-44000

2)Unclassified and A class road classes are lead to maximum number of accidents

3)Maximum number of accidents happened in the dry and wet/damp road surface
