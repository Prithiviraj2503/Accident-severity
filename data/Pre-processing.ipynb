```python
import pandas as pd

rawdata=pd.read_csv("accident.csv", encoding='cp1252')
rawdata.info()
```

```python
rawdata.head(3)
```

```python
rawdata['Type of Vehicle']=rawdata['Type of Vehicle'].replace(to_replace =['Bus or coach (17 or more passenger seats)','Minibus (8 – 16 passenger seats)'],  value ="Bus") 

rawdata['Type of Vehicle']=rawdata['Type of Vehicle'].replace(to_replace =['Car','Taxi/Private hire car'],  value ="car") 


rawdata['Type of Vehicle']=rawdata['Type of Vehicle'].replace(to_replace =['M/cycle 50cc and under','Motorcycle over 125cc and up to 500cc','Motorcycle over 50cc and up to 125cc','Motorcycle over 500cc'],  value ="Motarcycle") 

rawdata['Type of Vehicle']=rawdata['Type of Vehicle'].replace(to_replace =['Pedal cycle','Other Vehicle'],  value ="Pedal cycle") 


rawdata['Type of Vehicle']=rawdata['Type of Vehicle'].replace(to_replace =['Goods vehicle 3.5 tonnes mgw and under','Goods vehicle over 3.5 tonnes and under 7.5 tonnes mgw','Goods vehicle 7.5 tonnes mgw and over'],  value ="Goods Vehicle") 

rawdata['Type of Vehicle']=rawdata['Type of Vehicle'].replace(to_replace =['Agricultural vehicle (includes diggers etc.)'], value ="Agri Vehicle") 
```

```python
rawdata['Type of Vehicle'].unique()
```

## Drop Unwanted columns

```python
drpdata=rawdata.drop(['Reference Number','Easting','Northing','Accident Date','Time (24hr)'], axis=1)
```

## Encoding

```python
#Apply label Encoding for the ordinal feaures
from sklearn import preprocessing 

label_encoder = preprocessing.LabelEncoder() 
drpdata['1st Road Class']= label_encoder.fit_transform(drpdata['1st Road Class']) 
drpdata['Road Surface']= label_encoder.fit_transform(drpdata['Road Surface']) 
drpdata['Lighting Conditions']= label_encoder.fit_transform(drpdata['Lighting Conditions']) 
drpdata['Weather Conditions']= label_encoder.fit_transform(drpdata['Weather Conditions']) 
drpdata['Casualty Class']= label_encoder.fit_transform(drpdata['Casualty Class']) 
drpdata['Sex of Casualty']= label_encoder.fit_transform(drpdata['Sex of Casualty']) 
drpdata['Type of Vehicle']= label_encoder.fit_transform(drpdata['Type of Vehicle']) 
```
